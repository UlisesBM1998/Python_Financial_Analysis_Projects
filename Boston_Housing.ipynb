# IMPORTING LIBRARIES AND DATASET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
boston.keys()
```

```python
print(boston.DESCR)
```

```python
bostonData = boston.data
bostonFeat = boston.feature_names
bostonPrice = boston.target
```

# DATAFRAME

```python
boston = pd.DataFrame(data = bostonData, columns = bostonFeat)
boston
```

# STATISTICS FROM THE PRICE

```python
boston["Price"] = bostonPrice
```

```python
boston.describe()
```

```python
boston.describe().loc["mean"]
```

# CHECKING IN DATAFRAME FEATURE AND COUNT OF NAN, NULL DATA AND DUPLICATES

```python
boston.info()
```

```python
print(boston.isnull().sum())
print(boston.isna().sum())
```

```python
boston.duplicated().sum()
```

# DATA ANALYSIS

Getting the mean of all columns

```python
plt.style.use("ggplot")
boston.describe().loc["mean"].plot.bar()
```

Correlation plot of variables against all variables

```python
sns.pairplot(boston.iloc[:, np.hstack(([0], range(1,7)))], diag_kind = "kde")
```

```python
sns.pairplot(boston.iloc[:, np.hstack(([0], range(7,14)))], diag_kind = "kde")
```

Correlation plot with price variables vs. other variables

```python
rows = 2
cols = 7

fig,ax = plt.subplots(rows, cols, figsize = (16,4))

col = boston.columns
index = 0

for i in range(rows):
    for j in range(cols):
        sns.scatterplot(x = "Price", y = col[index], data = boston, ax= ax[i][j])
        index = index + 1
        
plt.tight_layout()
plt.show()
```

Correlation with a heat map

```python
fig, ax = plt.subplots(figsize = (10,10))
sns.heatmap(boston.corr(), annot = True, annot_kws = {"size": 10})
```

Boxplot comparing the price if the house is close to a River. We can see house are more expensive if they are close to this place.

```python
sns.boxplot( y= "Price", x = "CHAS", data = boston)
```

We can confirm that the price is correlated with the number of rooms and if the River is close.

```python
sns.relplot( x = "RM", y = "Price", data = boston, hue = "CHAS")
```

The same idea that the previous plot but identifying by density.

```python
sns.jointplot( x = "RM", y = "Price", data = boston, kind = "hex", color = "m")
```
